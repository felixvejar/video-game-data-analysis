```python
#Importar Librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st
```

```python
#Abrir el DataFrame a partir del archivo .csv
games = pd.read_csv("/datasets/games.csv")
games.info()
```

```python
#Pasar a minusculas los titulos de columna
games.columns = games.columns.str.lower()
games.info()
```

```python
#Revisar los valores nulos de la columna "name"
print(games[games["name"].isna()])
```

En las columnas "name" y "genre" en los valores nulos se podría colocar "Unknown", pues solamente dos juegos no cuentan con titulo y genero, pero si con la mayoria de los demás criterios. A las demás columnas podría colocarse 0 primero para las columnas que necesiten un cambio de objeto como en "year_of_release" y "user_score", se tendría que revisar si hay algun string oculto que impida el cambio de objeto.
No considero que afecten los valores nulos en las columnas de Score, ya que no afectarán a las operaciones aritméticas para el análisis posterior. 

```python
#Cambio de None a Unknown en la columna name
games["name"] = games["name"].fillna("Unknown")
#Cambio de None a Unknown en la columna genre
games["genre"] = games["genre"].fillna("Unknown")
#Rellenar valores nulos en la columna year_of_release
games["year_of_release"] = games["year_of_release"].fillna(0)
#Cambio de tipo de objeto en year_of_release
games["year_of_release"] = games["year_of_release"].astype(int)
#Cambio de 0 a Unknown en year_of_release
games["year_of_release"] = games["year_of_release"].replace(0, "Unknown")
#Checar el cambio en el resumen
games.info()
```

```python
#Comparación de XBox One y PC
#Filtrar por Xone
xone = games[(games["platform"] == "XOne")]
#Eliminar duplicados
xone_score = xone["user_score"].dropna()
#Filtrar por PC
pc = games[(games["platform"] == "PC")]
#Eliminar duplicados
pc_score = pc["user_score"].dropna()
#Prueba de Student
#H0: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
#H1: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC difieren.
alpha = 0.05
ttest_h0 = st.ttest_ind(xone_score, pc_score, equal_var = False)
#Resultado 
print("Valor p: ", ttest_h0.pvalue)
if ttest_h0.pvalue < alpha:
    print("Rechazamos H0")
    print("Por lo tanto, las calificaciones promedio para Xbox One y PC son diferentes")
else:
    print("No podemos rechazar H0")
    print("Por lo tanto, las calificaciones promedio para Xbox One y PC son similares")
```

```python
xone_mean = xone_score.mean()
print("Media User Score XBox One: ", xone_mean)
pc_mean = pc_score.mean()
print("Media User Score PC: ", pc_mean)
```
